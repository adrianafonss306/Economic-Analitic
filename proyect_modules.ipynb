```python
import json

# --- Funciones para abrir los archivos ---
def open_canasta():
    with open("./Canasta_Basica.json", "r", encoding="utf-8") as file:
        return json.load(file)

def open_mipymes():
    with open("./proyecto_icd.json", "r", encoding="utf-8") as file:
        return json.load(file)

# --- Procesar precios de alimentos y aseo ---
def process_products_food():
    data = open_mipymes()
    prices_food = {}
    for mipyme in data["mipymes"]:
        food = mipyme["store"]["products"][0]["groceries"]
        for producto, items in food.items():
            if producto not in prices_food:
                prices_food[producto] = []
            for k in items:
                if k["price"] not in ["", None]:
                    prices_food[producto].append(k["price"])
    return prices_food

def process_products_bath():
    data = open_mipymes()
    prices_bath = {}
    for mipyme in data["mipymes"]:
        bath = mipyme["store"]["products"][0]["hygiene"]
        for producto, items in bath.items():
            if producto not in prices_bath:
                prices_bath[producto] = []
            for k in items:
                if k["price"] not in ["", None]:
                    prices_bath[producto].append(k["price"])
    return prices_bath

# --- Promedios de todos los productos ---
def dict_average_all():
    prices_food = process_products_food()
    prices_bath = process_products_bath()
    all_prices = {**prices_food, **prices_bath}
    av_prices = {}
    for k, val in all_prices.items():
        if len(val) > 0:
            av_prices[k] = sum(val) / len(val)
    return av_prices

# --- Calcular costo por separado y total ---
def costo_canasta_separado():
    data = open_canasta()
    precios = dict_average_all()
    totales = {}

    for section in ["Canasta Basica", "Canasta Adicional"]:
        total_section = 0
        for producto, items in data[section].items():
            cantidad = sum(float(item["amount"]) for item in items if item["amount"] not in ["", None])
            if producto in precios:
                total_section += cantidad * precios[producto]
                print(f"{section} - {producto}: {cantidad}  {precios[producto]:.2f} = {cantidad * precios[producto]:.2f} CUP")
            else:
                print(f"{section} - {producto}: no encontrado en inventario")
        totales[section] = total_section

    print("\nCosto Canasta Bsica:", totales["Canasta Basica"], "CUP")
    print("Costo Canasta Adicional:", totales["Canasta Adicional"], "CUP")
    print("Costo Total:", totales["Canasta Basica"] + totales["Canasta Adicional"], "CUP")

    return totales

print(costo_canasta_separado())
```
