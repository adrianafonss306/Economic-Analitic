```python
def open_json():
    import json
    with open ("./proyecto_icd.json","r",encoding="utf-8") as file:
        data=json.load(file)
        
        return data
open_json()    
```

```python
def name_products():
    data = open_json()
    
    empty = []
    mipymes = data["mipymes"]
    
    for mipyme in range(len(mipymes)):
        products_food=mipymes[mipyme]["store"]["products"][0]["groceries"]         
        
        for j in products_food:
            if j not in empty:
                empty.append(j)
        products_bath=mipymes[mipyme]["store"]["products"][0]["hygiene"]
        
        for k in products_bath:  
            if k not in empty:
                empty.append(k)
    return empty  
print(name_products())
```

```python
def process_products_food():
    data = open_json()
    list_prices_food= {}
    count_food=0

    mipymes = data["mipymes"]
    for mipyme in range(len(mipymes)):
        for i in range(len(mipymes[mipyme]["store"]["products"][0])):
            food =mipymes[mipyme]["store"]["products"][0]["groceries"]
            for j, items in food.items(): 
                if j not in list_prices_food:
                    list_prices_food[j]=[] 
                for k in items:
                    count_food+=1
                    if k["price"] not in ["",None]:
                        list_prices_food[j].append(k["price"])
                        
    return list_prices_food

print(process_products_food())
            
```

```python
def load_data():
    """Carga los datos del JSON"""
    return open_json()

def process_product_items(items, list_prices, category_counter):
    """
    Procesa los items de productos y actualiza los precios
    """
    for product_name, product_list in items.items():
        if product_name not in list_prices:
            list_prices[product_name] = []
        
        for product in product_list:
            category_counter += 1
            if product["price"] not in ["", None]:
                list_prices[product_name].append(product["price"])
    
    return list_prices, category_counter



def process_mipyme_products(mipyme, list_prices, counters):
    """
    Procesa los productos de una mipyme individual
    """
    products = mipyme["store"]["products"][0]
    
    # Procesar productos de groceries
    list_prices, counters["food"] = process_product_items(
        products["groceries"], 
        list_prices, 
        counters["food"]
    )
    
    # Procesar productos de hygiene
    list_prices, counters["bath"] = process_product_items(
        products["hygiene"], 
        list_prices, 
        counters["bath"]
    )
    
    return list_prices, counters

def category_prices_refactored():
    """
    Obtiene los precios de todos los productos por categoría
    """
    data = load_data()
    list_prices = {}
    counters = {"food": 0, "bath": 0}
    
    for mipyme in data["mipymes"]:
        list_prices, counters = process_mipyme_products(
            mipyme, 
            list_prices, 
            counters
        )
    
    return list_prices

# Uso
print(category_prices_refactored())
```
